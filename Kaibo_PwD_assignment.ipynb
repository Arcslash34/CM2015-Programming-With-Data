<h3>Loading (Web scraping) and saving Country to continent data (Dataset 1)</h3>

For dataset 1, i used web scripting as the data I found is a table within the webpage, thus, i webscrape by locating the table with the table id, following which I saved it into a csv file for future use in case the website has gone down, table id or format of website has changed, which would render these code useless. Then, it is saved into a panda dataframe.

The file I have downloaded includes the countries and continents column which would be needed for me to combine the dataframe so that we would be able to compare the data accourding to the regions

```python
def countries_to_dataframe(url, output_csv_path='countries_data.csv'):
    try:
        # Fetch the HTML content from the provided URL using the global headers
        response = requests.get(url, headers=headers)
        
        # Check if the response status code is 200 (OK)
        response.raise_for_status()
        
        # Get the HTML content
        html_content = response.text
        
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(html_content, 'html.parser')

        # Find the table element by its id attribute
        table = soup.find(id='table_id')  # Replace 'table_id' with the actual ID
        
        # Process the found table into a pandas DataFrame
        if table:
            # Use pandas read_html to parse the table
            df = pd.read_html(str(table))[0]  # Read HTML table into DataFrame
            
            # Save DataFrame to CSV file
            df.to_csv(output_csv_path, index=False)
            print(f'Data saved to {output_csv_path}')
            
            return df
        else:
            print("Table not found. Trying to load from CSV...")
            if os.path.exists(output_csv_path):
                # Load data from CSV if it exists
                df = pd.read_csv(output_csv_path)
                print(f'Data loaded from {output_csv_path}')
                return df
            else:
                print(f'CSV file not found. Unable to retrieve data.')
                return None
    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")
        print(f'Trying to load data from CSV...')
        if os.path.exists(output_csv_path):
            # Load data from CSV if it exists
            df = pd.read_csv(output_csv_path)
            print(f'Data loaded from {output_csv_path}')
            return df
        else:
            print(f'CSV file not found. Unable to retrieve data.')
            return None

# Usage:
url = 'https://statisticstimes.com/geography/countries-by-continents.php'
df_continents = countries_to_dataframe(url)
if df_continents is not None:
    print(df_continents)
```

<h3>Loading (Web scraping) and saving GDP and elderly population data (Dataset 2 and 3)</h3>

Here for dataset 2 and 3, I used web scraping to locate the actual download link which have anchor (<code>&lt;a&gt;</code>) and the name csv, the reason would be that, after having this code, we would be able to just run the code and get the zip file instead of going to the actual website and downloading it all the time. Here I set the handle missing file function as the website would update the data frequently, rendering it to unable to locate the file within the zip file, thus I set it so it would refer to the previous file when running the code if the file is not found.<br>

The file downloaded below would be for gdp and elderly data with the countries and years. I would use this to compare with all the other data where we would be able to see demographic trends.

```python
# URL of the World Bank page
world_bank_url = "https://data.worldbank.org/indicator/NY.GDP.MKTP.CD"

# File paths and names
local_file_path2 = "gdp.zip"
target_file2 = "API_NY.GDP.MKTP.CD_DS2_en_csv_v2_6224532.csv"
# CSV file collected at 15/12/2023 12:06 GMT +8
previous_csv_file1 = "GDP.csv"

def handle_missing_file():
    print(f"{target_file2} not found. Using the previous CSV file.")
    if os.path.exists(previous_csv_file1):
        shutil.copy(previous_csv_file1, target_file2)
    else:
        print(f"The previous CSV file ({previous_csv_file1}) is missing.")

try:
    # Web scraping to find the download link
    response_world_bank = requests.get(world_bank_url, headers=headers)
    response_world_bank.raise_for_status()
    
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response_world_bank.text, 'html.parser')

    # Find a link containing the text "CSV"
    download_link_element = soup.find('a', string='CSV', href=True)

    if download_link_element:
        download_link = download_link_element['href']

        # Download the ZIP file
        response = requests.get(download_link)
        response.raise_for_status()
        
        with open(local_file_path2, 'wb') as file:
            file.write(response.content)

        print(f"ZIP file downloaded successfully to {local_file_path2}")

        # Extract and handle missing file
        with zipfile.ZipFile(local_file_path2, 'r') as zip_ref:
            (zip_ref.extract(target_file2) if target_file2 in zip_ref.namelist() else handle_missing_file())

        # Move the file to the desired name
        shutil.move(target_file2, "GDP.csv")
        print("GDP.csv loaded.")
    else:
        handle_missing_file()

except requests.RequestException as e:
    print(f"Failed to download ZIP file. Error: {e}")
except zipfile.BadZipFile:
    print("Invalid ZIP file. Unable to extract.")
except FileNotFoundError:
    handle_missing_file()
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

<h3>Loading (Web scripting) and saving birth rate data (Dataset 4)</h3>

Here I use web scripting with just the download link to download the zip file directly using the get method, this method is also better than going to the website and downloading it when needed. I choosed this method for dataset 4 as I am able to locate the actual download link to download directly instead of using the scraping method

The file downloaded below would be for birth rate data with the countries and years. I would use this to compare with all the other data where we would be able to see demographic trends.

```python
# URL of the CSV file collected at 15/12/2023 12:07 GMT +8
url = "https://api.worldbank.org/v2/en/indicator/SP.DYN.TFRT.IN?downloadformat=csv"

# Specify the local file path where you want to save the ZIP file
local_file_path4 = "birth_rate.zip"

# Specify the file you want to extract from the ZIP archive
target_file4 = "API_SP.DYN.TFRT.IN_DS2_EN_csv_v2_6224665.csv"

try:
    # Send a GET request to the URL
    response = requests.get(url)
    response.raise_for_status()  # Raise an HTTPError for bad responses

    # Get the content of the response
    content = response.content

    # Write the content to the local file
    with open(local_file_path4, 'wb') as file:
        file.write(content)

    print(f"ZIP file downloaded successfully to {local_file_path4}")

    # Extract only the specified file from the ZIP file
    with zipfile.ZipFile(local_file_path4, 'r') as zip_ref:
        if target_file4 in zip_ref.namelist():
            zip_ref.extract(target_file4)
            print(f"{target_file4} extracted successfully.")
        else:
            print(f"{target_file4} not found in the ZIP archive. Using the previously downloaded CSV file.")

    # If the extracted file is not found, use the original CSV downloaded previously
    if not os.path.exists(target_file4):
        print(f"Using the original CSV file: {target_file4}")
        # Assuming you have a backup file or the original CSV file in the same directory
        shutil.copy("birth_rate.csv", target_file4)

    # Move the file to the desired name
    shutil.move(target_file4, "birth_rate.csv")

except requests.RequestException as e:
    print(f"Failed to download ZIP file. Error: {e}")
except zipfile.BadZipFile:
    print("Invalid ZIP file. Unable to extract.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

Dual Scaling for Varied Magnitudes typically have significantly different magnitudes. Using dual axes accommodates these variations, preventing one variable from overshadowing the other. Enhanced Interpretability: Placing the data on separate axes enhances the interpretability of the visualization. This clarity is particularly valuable when communicating complex socioeconomic patterns to a diverse audience.

```python
def clear_html(text):
    with warnings.catch_warnings():
        warnings.simplefilter("ignore", category=UserWarning)
        soup = BeautifulSoup(text, 'html.parser')
    return soup.get_text()

def clear_markdown(text):
    # Remove * and **
    text = re.sub(r'\*{1,2}', '', text) 
    # Remove # and ## to ######
    text = re.sub(r'#*', '', text) 
    # Remove horizontal lines (---)
    text = re.sub(r'---', '', text) 
    return text

def export_text_from_notebook(input_notebook, output_text_file):
    with open(input_notebook, 'r', encoding='utf-8') as nb_file:
        notebook = nbformat.read(nb_file, as_version=4)
        
    # Extract code and markdown
    relevant_content = []
    for cell in notebook['cells']:
        if cell['cell_type'] in ['markdown']:
            cell_content = clear_html(cell['source'])
            cell_content = clear_markdown(cell_content)
            relevant_content.append(cell_content)
            
    # Export as txt file
    with open(output_text_file, 'w', encoding='utf-8') as txt_file:
        txt_file.write('\n'.join(relevant_content))
```

```python
export_text_from_notebook("Kaibo_PwD_assignment.ipynb", "pwd_data.txt")
```

```python
import nltk
from nltk import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from wordcloud import WordCloud
from nltk.stem import WordNetLemmatizer
from PIL import Image

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
# Read the data
with open('pwd_data.txt', 'r', encoding='utf-8') as f:
    data = f.read()

# Tokenize the data
tokens = word_tokenize(data)

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize tokens
lemma_tokens = [lemmatizer.lemmatize(word.lower()) for word in tokens]

# Load stop words
stop_words = set(stopwords.words('english'))

# Add custom stop words
stop_words.update({"however"})

# Filter tokens
filtered_data = [w for w in lemma_tokens if w.lower() not in stop_words and w.isalpha()]

# Output
print("Filtered Data:", filtered_data)
```

```python
plt.figure(figsize = (12,4))
nltk.FreqDist(filtered_data).plot(12, cumulative=False, title="Word Plot")
plt.show()
```

```python
# Create a frequency distribution
counted_data = FreqDist(filtered_data)

# Using image shape for the word cloud
world_mask = np.array(Image.open("world_shape.jpg"))

# Create a word cloud from frequencies
wordcloud = WordCloud(
    width = 900,
    height = 450,
    background_color = "white",
    max_words = 45,
    relative_scaling = 0.4,
    mask = world_mask,
    colormap = "rainbow",
    contour_width = 0.4,
    contour_color='white'
).generate_from_frequencies(counted_data)

# Display the word cloud
plt.figure(figsize = (10,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
